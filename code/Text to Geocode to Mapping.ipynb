```python
import pandas as pd
import requests
from xml.etree import ElementTree
import numpy as np
import folium
from folium.plugins import MarkerCluster
```

```python
GetAllCountersUrl = "http://webservices.commuterpage.com/counters.cfc?wsdl&method=GetAllCounters"
xmlfile = open('xml_getallcounters.xml', 'w')
xmldata = requests.get(GetAllCountersUrl)
xmlfile.write(xmldata.text)
xmlfile.close()
xml_data = 'xml_getallcounters.xml'
```

```python
tree = ElementTree.parse(xml_data)
counter = tree.find('counter')
name = counter.find('name')
name.text
```

```python
counter.find('latitude').text
```

```python
counter.find('longitude').text
```

```python
counter.find('region/name').text
```

```python
list_of_dfs = []
for c in tree.findall('counter'):
    
    df_counters = pd.DataFrame(
       {'ID' : c.attrib['id'],
        'Name' : c.find('name').text,
        'latitude' : c.find('latitude').text,
        'longitude' : c.find('longitude').text,
        'region' : c.find('region/name').text
       }, index = [0])
    list_of_dfs.append(df_counters)
df = pd.concat(list_of_dfs)
df.head()
```

```python
df[['Name', 'latitude', 'longitude']][3]
```

```python
locations = df[['latitude', 'longitude']]
locationlist = locations.values.tolist()
name = df[['Name']]
namelist = name.values.tolist()
namelist[:3]
```

```python
map = folium.Map(location=[38.9, -77.05], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=namelist[point]).add_to(map)
map
```

```python
map2 = folium.Map(location=[38.9, -77.05], tiles='CartoDB dark_matter', zoom_start=11)
marker_cluster = folium.plugins.MarkerCluster().add_to(map2)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=namelist[point]).add_to(marker_cluster)
map2
```

```python
map2 = folium.Map(location=[38.9, -77.05], tiles='Stamen Terrain', zoom_start=11)
marker_cluster = folium.plugins.MarkerCluster().add_to(map2)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=namelist[point], icon=folium.Icon(color='darkblue', icon_color='white', icon='male', angle=0, prefix='fa')).add_to(marker_cluster)
map2
```

```python
def regioncolors(counter):
    if counter['region'] == 'Arlington':
        return 'green'
    elif counter['region'] == 'Alexandria':
        return 'blue'
    elif counter['region'] == 'DC':
        return 'red'
    else:
        return 'darkblue'
    
df["color"] = df.apply(regioncolors, axis=1)
df.head()
```

```python
color_list = df[['color']].values.tolist()
color_list[:3]
```

```python
map3 = folium.Map(location=[38.9, -77.05], tiles='CartoDB positron', zoom_start=11)
marker_cluster = folium.plugins.MarkerCluster().add_to(map3)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup='ID:'+df_counters['ID'][point]+' '+df_counters['Name'][point], icon=folium.Icon(color=df_counters["color"][point], icon_color='white', icon='male', angle=0, prefix='fa')).add_to(marker_cluster)
map3
```

```python
map3 = folium.Map(location=[38.9, -77.05], tiles='CartoDB positron', zoom_start=11)
marker_cluster = folium.plugins.MarkerCluster().add_to(map3)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], 
                  popup=namelist[point], 
                  icon=folium.Icon(color=color_list[point], 
                                   icon_color='white', 
                                   icon='male', 
                                   angle=0, 
                                   prefix='fa')).add_to(marker_cluster)
map3
```
